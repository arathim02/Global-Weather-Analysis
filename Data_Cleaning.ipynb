# 4. Standardize text columns

```python
text_cols = ['country', 'location_name', 'timezone', 'condition_text', 
             'wind_direction', 'moon_phase']

for col in text_cols:
    df[col] = df[col].str.strip().str.title()
```

```python
df
```

# 5. Remove impossible values


## Temperature check

```python
df = df[(df['temperature_celsius'] > -60) & (df['temperature_celsius'] < 60)]
```

```python
df
```

### Humidity should be 0 to 100


```python
df = df[(df['humidity'] >= 0) & (df['humidity'] <= 100)]
```

```python
df
```

#### Cloud should be 0 to 100

```python
df = df[(df['cloud'] >= 0) & (df['cloud'] <= 100)]
```

```python
df
```

##### Wind degree 0 to 360

```python
df = df[(df['wind_degree'] >= 0) & (df['wind_degree'] <= 360)]
```

```python
df
```

##### UV index positive

```python
df = df[df['uv_index'] >= 0]
```

```python
df
```

##### 6. Drop unnecessary columns 

```python
df = df.drop(columns=['last_updated_epoch'])

```

```python
df
```

# year

```python
df['last_updated'] = pd.to_datetime(df['last_updated'])
df['year'] = df['last_updated'].dt.year
df['month'] = df['last_updated'].dt.month
df['hour'] = df['last_updated'].dt.hour
```
